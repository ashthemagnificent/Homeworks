#Exploring and Visualizing Data in Python

#### Loading Data

```python
# import required libraries for data visualization and analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import *
import seaborn as sns
```

```python
# Load data from file "adult.csv" as dataframe
df = pd.read_csv('/content/adult.csv')
```

#### Exploring Data Structure

```python
# Show summary of records. Print names of columns and their datatypes
df.info()
```

```python
# Check your data looking with first 5 rows of table. Print the size of data
print(df.head(5))
df.shape
```

#### Exploring Data Statictically

```python
# Get statistical summary of data
summary = df.describe(include = 'all')
summary
```

```python
# Find mean values of the first 10 records in dataframe
df.head(10).select_dtypes(include='number').mean()
```

#### Filtering Data

```python
# Get records whose "age" is greater and equal than 25
older_25 = df[df['age']>=25]
older_25
```

```python
# Obtain a subspace of data with "Age","Gender" and "income" features
subspace= df[["age","gender","income"]]
subspace
```

#### Data Cleaning

```python
# Check duplicate records
df1 = df.drop_duplicates(keep='first')
df1
```

```python
# Convert categorical "martial-status" data to binary attibutes
binary_marital_status = pd.get_dummies(df1['marital-status'])
binary_marital_status
```

```python
# Check missing values in your data
print(df1.isnull().sum())
df1[df1 == '?'].count()
```

```python
# Obtain a subset of data with nonmising values
df2 = df1.replace('?', pd.NA).copy()
df2.dropna(inplace=True)
df2
```

```python
# Fill missing values using any technique (mentioned in the lecture) according to data type
df3 = df1.replace('?', pd.NA).copy()
mode_values = df1.mode().iloc[0]
df3 = df3.fillna(mode_values)
df3
```

#### Visualizing Data

1.Univariate analysis

```python
# "Age" distibution using histogram, pdf and cdf (using matplotlib  and seaborn)
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(30, 8))
sns.histplot(df3['age'], bins=25, ax=axes[0])
axes[0].set_title('Age Distribution (Histogram)')
sns.histplot(df3['age'], bins=25, kde= True, stat="density", ax=axes[1])
axes[1].set_title('Age Distribution (PDF)')
sns.histplot(df3['age'], bins=25, kde= True, cumulative=True, stat="density", common_norm=False, ax=axes[2])
axes[2].set_title('Age Distribution (CDF)')
plt.tight_layout()
plt.show()
```

```python
# Count plot for categorical attibute "education","martial-status" and "occupation" (using seaborn)
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(18, 12))
sns.countplot(x='education', data=df3, ax=axes[0], order=df3['education'].value_counts().index)
axes[0].set_title('Count Plot for Education')
sns.countplot(x='marital-status', data=df3, ax=axes[1], order=df3['marital-status'].value_counts().index)
axes[1].set_title('Count Plot for Martial Status')
sns.countplot(x='occupation', data=df3, ax=axes[2], order=df3['occupation'].value_counts().index)
axes[2].set_title('Count Plot for Occupation')
plt.tight_layout()
plt.show()
```

```python
# Draw violin plot for "educational-num" feature
plt.figure(figsize=(8, 6))
sns.violinplot(x='educational-num', data=df3)
plt.title('Violin Plot for Educational-Num')
plt.show()
```

2.Bivariate analysis

```python
# Draw Seaborn scatter plot of "age" and "fnlwgt" features (colored according to classes (hint:hue))
plt.figure(figsize=(10, 10))
sns.scatterplot(x='age', y='fnlwgt', hue='workclass', data=df3)
plt.title('Scatter Plot of Age vs. Fnlwgt (Colored by workclass)')
plt.show()
```

3.Multivariate analysis

```python
# Calculate correlation of features and show using heatmap of Seaborn.
numeric_columns = df3.select_dtypes(include='number')
correlation_matrix = numeric_columns.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

 Try to interpret plot

In our correlation heatmap, the darker and more opaque the color, the higher the correlation is considered. Darker red indicates a stronger positive correlation, while darker blue indicates a negative correlation. In our dataset, no strong correlation is observed among the categories, as all values are <0.2. Therefore, we can conclude that there is no strong correlation.

```python
# Draw seaborn pair plot demonstrating class info.
sns.pairplot(df3, hue='workclass')
plt.suptitle('Pair Plot with Class Information', y=1.02)
plt.show()
```

 Try to interpret plot

We observe that there is no significant correlation between pairs. However, it is noticeable that capital gain values only differ from zero when capital loss is zero. The general tendency is for capital gain values to be below 50000, with the upper limit around 99000 and no values in between. Analyzing the graph with the fnlwgt value, we can see that capital gain surpasses 50000 only when fnlwgt is below 0.75. It is evident that, at almost every age and for nearly every hours per week exceeding 20, capital gain can exceed 50000. Additionally, the person with the highest hours per week and capital gain values is in the 'self-emp-inc' workclass.
